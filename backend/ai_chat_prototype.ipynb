# AI Chat Prototype
This notebook demonstrates how to call the local AI endpoint (`/ai/ask`) to retrieve the same user-specific stats the server returns.

- Update `server_url` and `auth_token` below as needed.
- The example sends a natural question (e.g. `Show my personal stats`) and the server's agent will call the tool and return a cleaned answer.

```python
# Install dependencies if you don't have them (run once)
# !pip install requests python-dotenv
import os
import requests
from pprint import pprint
```

```python
# Configuration - point to your running backend and (optionally) provide a JWT auth token
server_url = os.getenv('AI_SERVER', 'http://localhost:3000')
endpoint = f'{server_url}/ai/ask'
# If your dev server expects an Authorization header, set it here (e.g. 'Bearer <token>')
auth_token = os.getenv('AI_AUTH_TOKEN', '')
headers = {'Content-Type': 'application/json'}
if auth_token:
    headers['Authorization'] = auth_token
```

```python
def ask_ai(question: str, timeout: int = 30):
    payload = {'question': question}
    r = requests.post(endpoint, json=payload, headers=headers, timeout=timeout)
    r.raise_for_status()
    return r.json()

# Example: request personal stats (the server agent should call the tool internally)
resp = ask_ai('Show my personal stats')
pprint(resp)
```
